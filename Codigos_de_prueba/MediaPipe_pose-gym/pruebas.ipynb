```python
# Importar dependencias
import cv2
import mediapipe as mp
import numpy as np
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
# Determinando los puntos de las articulaciones
# Reemplaza con la ruta de tu archivo de video
video_path = 'C:/Users/USUARIO/Proyectos_IA/Datasets/videos_gym/press_incorrectos/press_inclinado_cristian_incorrecto.mp4'
cap = cv2.VideoCapture(video_path)

# cap = cv2.VideoCapture(0)

# Obtener y mostrar el FPS del video original
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"FPS del video original: {fps}")

# Configurar instancia de mediapipe
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # Cambiar el color de la imagen a RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
      
        # Hacer deteccin
        results = pose.process(image)
    
        # Cambiar el color de nuevo a BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Redimensionar el video a 640x640 pxeles
        image = cv2.resize(image, (570, 800))
        
        # Extraer puntos de referencia
        try:
            landmarks = results.pose_landmarks.landmark
            print(landmarks)
        except:
            pass
        
        
        # Detecciones de renderizado
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                                mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2) 
                                 )               
        
        cv2.imshow('Mediapipe Feed', image)

        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
# Calcular ngulos para el press de banca o para las piernas
def calculate_joint_angle(pointA, pointB, pointC):
    """
    Calcula el ngulo formado por tres puntos (A, B, C).
    pointA: Coordenadas del primer punto (por ejemplo, hombro o cadera)
    pointB: Coordenadas del segundo punto (por ejemplo, codo o rodilla)
    pointC: Coordenadas del tercer punto (por ejemplo, mueca o tobillo)
    """
    
    # Convertir los puntos a arrays de NumPy
    pointA = np.array(pointA)  # Primer punto (A)
    pointB = np.array(pointB)  # Punto medio (B)
    pointC = np.array(pointC)  # Punto final (C)
    
    # Calcular el ngulo en radianes
    radians = np.arctan2(pointC[1] - pointB[1], pointC[0] - pointB[0]) - np.arctan2(pointA[1] - pointB[1], pointA[0] - pointB[0])
    angle = np.abs(radians * 180.0 / np.pi)  # Convertir a grados
    
    # Asegurarse de que el ngulo est dentro del rango de 0-180 grados
    if angle > 180.0:
        angle = 360.0 - angle
        
    return angle
```

```python
# Contador de repeticiones para press de banca
# Reemplaza con la ruta de tu archivo de video
video_path = 'C:/Users/USUARIO/Proyectos_IA/Datasets/videos_gym/press_incorrectos/press_plano_cristian_incorrecto.mp4'
cap = cv2.VideoCapture(video_path)

# Video webcam
# cap = cv2.VideoCapture(0)

# Variables del contador
counter = 0 
stage = None

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        if not ret:
            break
        
        # Cambiar el color de la imagen a RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # Hacer deteccin
        results = pose.process(image)
        
        # Cambiar el color de nuevo a BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Redimensionar el video a 0x0 pxeles
        image = cv2.resize(image, (570, 800))
        
        # Extraer puntos de referencia
        try:
            landmarks = results.pose_landmarks.landmark
            
            # Obtener coordenadas del brazo izquierdo
            left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                             landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                          landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                          landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            
            # Obtener coordenadas del brazo derecho
            right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
            right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
            
            # Obtener coordenadas de la pierna izquierda
            left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                        landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                          landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]

            # Obtener coordenadas de la pierna derecha
            right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                         landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
            right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
            
            # Calcular ngulos de los brazos
            left_arm_angle = calculate_joint_angle(left_shoulder, left_elbow, left_wrist)
            right_arm_angle = calculate_joint_angle(right_shoulder, right_elbow, right_wrist)
            
            # Calcular ngulos de las piernas
            left_leg_angle = calculate_joint_angle(left_hip, left_knee, left_ankle)
            right_leg_angle = calculate_joint_angle(right_hip, right_knee, right_ankle)
            
            # Verificar y corregir la ejecucin del ejercicio
            correction_message = ""
            correction_details = []
            corrected = False
            
            # Correccin para los brazos
            if abs(left_arm_angle - right_arm_angle) > 15:
                correction_details.append("Corrige la simetria de los brazos.")
                corrected = True
            
            if not (90 <= left_arm_angle <= 120):
                correction_details.append("Extiende completamente el brazo izquierdo al subir.")
                corrected = True
            if not (90 <= right_arm_angle <= 120):
                correction_details.append("Extiende completamente el brazo derecho al subir.")
                corrected = True
            
            if left_arm_angle < 50:
                correction_details.append("Flexiona adecuadamente el brazo izquierdo al bajar.")
                corrected = True
            if right_arm_angle < 50:
                correction_details.append("Flexiona adecuadamente el brazo derecho al bajar.")
                corrected = True
            
            # Correccin para las piernas
            if not (80 <= left_leg_angle <= 120):
                correction_details.append("Ajusta la posicion de tu pierna izquierda.")
                corrected = True
            if not (20 <= right_leg_angle <= 90):
                correction_details.append("Ajusta la posicion de tu pierna derecha.")
                corrected = True
            
            # Mostrar mensajes de correccin
            if corrected:
                y0, dy = 50, 20
                for i, line in enumerate(correction_details):
                    y = y0 + i * dy
                    cv2.putText(image, line, (image.shape[1] - 300, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2, cv2.LINE_AA)
            
            # Mostrar mensajes de simetra de los brazos
            if left_arm_angle > right_arm_angle + 15:  # Si el brazo izquierdo est ms alto
                cv2.putText(image, "Brazo izquierdo mas alto", (image.shape[1] - 300, y0 + len(correction_details) * dy + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
            elif right_arm_angle > left_arm_angle + 15:  # Si el brazo derecho est ms alto
                cv2.putText(image, "Brazo derecho mas alto", (image.shape[1] - 300, y0 + len(correction_details) * dy + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
            
            # Lgica del contador de repeticiones para press de banca
            if left_arm_angle > 100 and right_arm_angle > 100:  # Ambos brazos extendidos (parte superior)
                stage = "up"
            if left_arm_angle < 50 and right_arm_angle < 50 and stage == 'up':  # Ambos brazos en ngulo de 45 grados (parte inferior)
                stage = "down"
                counter += 1
                print(counter)
            
            # Visualizar ngulos
            left_elbow_coords = tuple(np.multiply(left_elbow, [image.shape[1], image.shape[0]]).astype(int))
            right_elbow_coords = tuple(np.multiply(right_elbow, [image.shape[1], image.shape[0]]).astype(int))
            
            cv2.putText(image, str(int(left_arm_angle)), 
                        left_elbow_coords, 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            
            cv2.putText(image, str(int(right_arm_angle)), 
                        right_elbow_coords, 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

            left_knee_coords = tuple(np.multiply(left_knee, [image.shape[1], image.shape[0]]).astype(int))
            right_knee_coords = tuple(np.multiply(right_knee, [image.shape[1], image.shape[0]]).astype(int))

            cv2.putText(image, str(int(left_leg_angle)), 
                        left_knee_coords, 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            
            cv2.putText(image, str(int(right_leg_angle)), 
                        right_knee_coords, 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

        except Exception as e:
            print(f"Error: {e}")
        
        # Renderizar contador de repeticiones
        # Cuadro de estado de configuracin
        cv2.rectangle(image, (0, 0), (225, 73), (245, 117, 16), -1)
        
        # Mostrar repeticiones
        cv2.putText(image, 'REPS', (15, 12), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter), 
                    (10, 60), 
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
        
        # Mostrar etapa del movimiento
        cv2.putText(image, 'STAGE', (65, 12), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, stage, 
                    (60, 60), 
                    cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
        
        # Detecciones de renderizado
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2), 
                                mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))               
        
        cv2.imshow('Mediapipe Feed', image)

        if cv2.waitKey(5) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
